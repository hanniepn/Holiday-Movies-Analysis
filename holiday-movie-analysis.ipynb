## Processing and Cleaning Datasets

```python
import pandas as pd
```

```python
# Read Dataset 1
df1 = pd.read_csv('/Users/hanniepham/Documents/Data200 Final/holiday_movies.csv')
```

```python
df1
```

```python
# Check dataset 1 information
print(df1.head())  
print(df1.info())
```

```python
# Clean dataset 1 by checking to see if there is any missing value in dataset
missing_values = df1.isnull().sum()
print(missing_values)
```

```python
# Fill up missing values in genres column
most_frequent_category = df1['genres'].mode()[0]
df1['genres'].fillna(most_frequent_category, inplace=True)
```

```python
# Fill up missing values in runtime_minutes column
df1['runtime_minutes'].fillna(df1['runtime_minutes'].mean(), inplace=True)
```

```python
# Double-check missing values
missing_values = df1.isnull().sum()
print(missing_values)
```

```python
# Export to new cleaned dataset
df1.to_csv('cldataset1.csv', index=False)
```

```python
# Read Dataset 2
df2 = pd.read_csv('/Users/hanniepham/Documents/Data200 Final/movies2.csv')
df2
```

```python
# Check dataset 2 information
print(df2.head())  
print(df2.info())
```

```python
# Clean dataset 2 by checking to see if there is any missing value in dataset 
missing_values = df2.isnull().sum()
print(missing_values)
```

```python
# Fill up missing values in categorical columns
categorical_columns = df2.select_dtypes(include=['object']).columns
for column in categorical_columns:
    most_frequent_category = df2[column].mode()[0]
    df2[column].fillna(most_frequent_category, inplace=True)
```

```python
# Fill up missing values in numerical columns
numerical_columns = df2.select_dtypes(include=['int64', 'float64']).columns
for column in numerical_columns:
    df2[column].fillna(df2[column].mean(), inplace=True)
```

```python
# Double-check missing values
missing_values = df2.isnull().sum()
print(missing_values)
```

```python
# Export to new cleaned dataset
df2.to_csv('cldataset2.csv', index=False)
```
