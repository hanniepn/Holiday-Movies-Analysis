## Processing and Cleaning Datasets

```python
import pandas as pd
```

```python
# Read Dataset 1
df1 = pd.read_csv('/Users/hanniepham/Documents/Data200 Final/holiday_movies.csv')
```

```python
df1
```

```python
# Check dataset 1 information
print(df1.head())  
print(df1.info())
```

```python
# Clean dataset 1 by checking to see if there is any missing value in dataset
missing_values = df1.isnull().sum()
print(missing_values)
```

```python
# Fill up missing values in genres column
most_frequent_category = df1['genres'].mode()[0]
df1['genres'].fillna(most_frequent_category, inplace=True)
```

```python
# Fill up missing values in runtime_minutes column
df1['runtime_minutes'].fillna(df1['runtime_minutes'].mean(), inplace=True)
```

```python
# Double-check missing values
missing_values = df1.isnull().sum()
print(missing_values)
```

```python
# Export to new cleaned dataset
df1.to_csv('cldataset1.csv', index=False)
```

```python
# Read Dataset 2
df2 = pd.read_csv('/Users/hanniepham/Documents/Data200 Final/movies2.csv')
df2
```

```python
# Check dataset 2 information
print(df2.head())  
print(df2.info())
```

```python
# Clean dataset 2 by checking to see if there is any missing value in dataset 
missing_values = df2.isnull().sum()
print(missing_values)
```

```python
# Fill up missing values in categorical columns
categorical_columns = df2.select_dtypes(include=['object']).columns
for column in categorical_columns:
    most_frequent_category = df2[column].mode()[0]
    df2[column].fillna(most_frequent_category, inplace=True)
```

```python
# Fill up missing values in numerical columns
numerical_columns = df2.select_dtypes(include=['int64', 'float64']).columns
for column in numerical_columns:
    df2[column].fillna(df2[column].mean(), inplace=True)
```

```python
# Double-check missing values
missing_values = df2.isnull().sum()
print(missing_values)
```

```python
# Export to new cleaned dataset
df2.to_csv('cldataset2.csv', index=False)
```

### Visualization 1

According to the visual 1 page, the visualization "Company Trends Over Years" represents an in-depth examination of the film production industry over a thirty-year period from 1986 to 2016. It provides insight into two key metrics for film production companies: the count of years in which they have produced movies and the average score of the movies they produced. This dual-axis chart combines a bar graph, representing the count of years, with a line graph that depicts the average movie scores.

The bar graph indicates the number of years companies have been actively creating movies, which is a clear reflection of their industry longevity and operational stability. Companies that have set the highest standards, such as Columbia Pictures and Paramount, have demonstrated a large and continuous presence throughout a three-decade period. This persistent production implies a steady market position, durability in the ever-changing film industry, and possibly a robust film portfolio.
The gradient in bar heights observed from left to right suggests that, while some organizations have a lengthy and continuous production history, others have been less consistent or have a shorter period of activity. This could be due to varied corporate strategies, market entrance points, or the long-term influence of industry developments. The range of years also reflects differing dimensions of operation and methods to market participation, with some companies focused on volume and others on specialized markets or selective manufacture.

Overlaying the production volume is the line graph indicating the average scores of movies produced by these companies. The scores likely come from aggregators like IMDb, reflecting critical reception and audience appreciation. The average score is an indicator of the perceived quality of a company's filmography. An intriguing finding is that companies with the biggest production volume do not always have the highest average scores. This could imply that a focus on quantity does not always correspond to quality in the opinion of viewers and critics. In contrast, some firms with fewer active years had higher average scores, which could indicate a strategy centered on producing fewer but higher-quality films that engage strongly with consumers and critics.

The relationship between the count of years and average scores is not linear, suggesting that longevity and productivity do not uniformly result in higher movie scores. This non-correlation can have multiple implications:

- Market Saturation: Companies with a high volume of productions may saturate the market, potentially diluting the average quality of their offerings.
- Selective Production: Companies with fewer productions but higher average scores might focus on carefully curated projects, which could attract a particular audience or cater to a specific genre with dedicated followers.
- Strategic Positioning: Companies with both high production counts and above-average scores may have successfully found a balance, possibly through a mix of blockbuster and niche films, or through strong quality control mechanisms despite high output.

In conclusion, the "Company Trends Over Years" visualization offers a valuable snapshot of the film production industry from 1986 to 2016, highlighting the varying strategies of companies regarding their production volume and the quality of their films. It reveals that while some companies have focused on maintaining a high output over the years, others have prioritized the quality of their films, as evidenced by their average scores. This data provides a foundation for understanding the complexities of the film production industry and suggests that a company's success cannot be measured by longevity and productivity alone, but also by the lasting impact of its films as reflected in their scores.
